### __2. Check then show if there are any players with the name `ronaldo`__

You should check if any of the values in the column 'player' contain the string 'ronaldo'

__Note__: your solution should be case insensitive, meaning that it should consider "RONaldo" as a valid `ronaldo` case

image.png

```python
# Answer here
# Checking and showing if there are any players with the name 'ronaldo' (case insensitive)
ronaldo_players = df[df['player'].str.contains('ronaldo', case=False, na=False)]
ronaldo_players

```

### __3. Get the two rows of the players `Lionel Messi` and `Cristiano Ronaldo`.__

Extract the rows that contain these players into a separate dataframe

image.png

```python
# Answer here
# Getting the rows for the players Lionel Messi and Cristiano Ronaldo
messi_ronaldo_players = df[df['player'].str.contains('Lionel Messi|Cristiano Ronaldo', case=False, na=False)]
messi_ronaldo_players

```

### __4. Compare the information of these two players `Lionel Messi` and `Cristiano Ronaldo` using a plot.__

plot all the columns
`columns_to_plot = ['height', 'weight', 'age', 'ball_control', 'dribbling','slide_tackle', 'gk_kicking']`
into a plot.

The x axis should be the Attributes and the y axis should be the values of these columns.

The end result should look like this:


image.png

```python
# Answer here
import numpy as np

# Selecting the columns to plot
columns_to_plot = ['height', 'weight', 'age', 'ball_control', 'dribbling', 'slide_tackle', 'gk_kicking']

# Extracting the relevant data for Messi and Ronaldo
messi_ronaldo_data = messi_ronaldo_players[columns_to_plot]

# Transposing the DataFrame for plotting
messi_ronaldo_transposed = messi_ronaldo_data.T

# Creating the plot
plt.figure(figsize=(12, 8))
for player in messi_ronaldo_transposed.columns:
    plt.plot(messi_ronaldo_transposed.index, messi_ronaldo_transposed[player], marker='o', label=player)

plt.title('Comparison of Lionel Messi and Cristiano Ronaldo')
plt.xlabel('Attributes')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

```
